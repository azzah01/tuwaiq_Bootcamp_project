```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

```python

#plt.style.use('_mpl-gallery')



df = pd.read_csv("healthcare-dataset-stroke-data.csv")

df= df.drop(df[df["gender"]=="Other"].index)



print(df.tail(5))

print(df.info())



```

```python


#sns.catplot(x="stroke", y="age", data=selected)


selected = df[["age", "stroke", "gender"]]
sns.catplot(x="gender", y="age", hue="stroke", kind="bar", data=df)
plt.savefig('./gender3.jpg')
plt.show()

```

```python

sns.catplot(x="stroke", y="age", hue="ever_married", kind="box", data=df)

plt.show()

sns.catplot(x="stroke", y="age", data=df)
plt.show()


sns.catplot(x="age", y="smoking_status" , hue="stroke",kind="violin",  data=df)
plt.show()

```

```python

sns.catplot(x="age", y="smoking_status" , hue="stroke",kind="bar",  data=df)

plt.show()

```

```python
for c in range(1,11):

    sns.catplot(x="stroke", y=df.columns[c], kind="bar", data=df)
    plt.savefig("./"+df.columns[c]+".png")
    plt.show()

exit()
```
